```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("C:\\Users\\Rishu Jaiswal\\Downloads\\archive (2)\\Data\\features_30_sec.csv")
```

```python
feature_column = 'spectral_centroid_mean'

# Extract the feature data
feature_data = df[feature_column]

# Calculate mean and variance
mean = np.mean(feature_data)
variance = np.var(feature_data)



# Plot histogram
a = plt.hist(feature_data, bins=10)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of {}'.format(feature_column))
plt.show()

print("Mean:", mean)
print("Variance:", variance)
```
