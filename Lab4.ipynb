```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#finding number of classes
np.unique(df['label'])
```

```python
#dataframe for blues class
blues_class_df = df[df['label']=='blues']

blues_class_df
```

```python
#mean(class centroid) for blues class
blues_class_mean = np.mean(blues_class_df.iloc[:,1:-1], axis=0)
blues_class_mean
```

```python
#Spread(standard deviation) for blues class
blues_class_std = np.std(blues_class_df.iloc[:,1:-1], axis = 0)
blues_class_std
```

```python
#dataframe for classical class
classical_class_df = df[df['label']=='classical']
```

```python
#mean(centroid) for classical class
classical_class_mean = np.mean(classical_class_df.iloc[:,1:-1], axis=0)
classical_class_mean
```

```python
#spread(standard deviation) for classical class
classical_class_std = np.std(classical_class_df.iloc[:,1:-1], axis=0)
classical_class_std
```

```python
#distance between mean vectors between class blues and class classical
np.linalg.norm(blues_class_mean - classical_class_mean)
```
