```python
# Assuming you have two classes represented by their feature vectors
class1_data = data['rms_mean']  
class2_data = data['rms_var']  

# Calculate the mean (centroid) for each class
centroid_class1 = np.mean(class1_data)
centroid_class2 = np.mean(class2_data)

# Calculate the spread (standard deviation) for each class
spread_class1 = np.std(class1_data)
spread_class2 = np.std(class2_data)

# Calculate the distance between mean vectors of the two classes
distance_between_centroids = np.linalg.norm(centroid_class1 - centroid_class2)

# Print results
print("Mean (centroid) of class 1:", centroid_class1)
print("Mean (centroid) of class 2:", centroid_class2)
print("Spread (standard deviation) of class 1:", spread_class1)
print("Spread (standard deviation) of class 2:", spread_class2)
print("Distance between centroids of class 1 and class 2:", distance_between_centroids)

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming 'class1_data' and 'class2_data' are NumPy arrays containing the data
class1_data = np.array(class1_data)
class2_data = np.array(class2_data)

# Calculate centroids
centroid_class1 = np.mean(class1_data)
centroid_class2 = np.mean(class2_data)

# Calculate spreads
spread_class1 = np.std(class1_data)
spread_class2 = np.std(class2_data)

# Calculate distance between centroids
distance_between_centroids = np.linalg.norm(centroid_class1 - centroid_class2)

# Plotting
plt.scatter(range(len(class1_data)), class1_data, label='Class 1 Data')
plt.scatter(range(len(class2_data)), class2_data, label='Class 2 Data')

plt.errorbar(centroid_class1, centroid_class1, yerr=spread_class1, fmt='r*', label='Class 1 Centroid')
plt.errorbar(centroid_class2, centroid_class2, yerr=spread_class2, fmt='r*', label='Class 2 Centroid')

# Add text for distance between centroids
plt.text((centroid_class1 + centroid_class2) / 2, min(centroid_class1, centroid_class2), f'Distance: {distance_between_centroids:.2f}', ha='center', color='blue')

plt.xlabel('Samples')
plt.ylabel('Feature Values')
plt.title('Dataset with Centroids and Spread')
plt.legend()
plt.grid(True)
plt.show()

```
